3. Analysis Questions

*   Average final grade (G3).
*   How many students scored above 15?
*   Is study time correlated with performance?
*   Which gender performs better on average?

```python
print("Average Final Grade:", df['G3'].mean())
```

```python
above_fifteen= df[df['G3'] > 15] #to see the student data scored above 15?
above_15= df[df['G3'] > 15].shape[0]
print("Number of students scored above 15:\n", above_15)
```

```python
correlation = df["studytime"].corr(df["G3"])
print("study time correlated with performance", correlation)
```

```python
gen_avg= df.groupby('sex')['G3'].mean()
print("gender performs better on average:-\n", gen_avg)
```

4. Visualizations

*   Histogram of grades.
*   Scatterplot: study time vs grades.
*   Bar chart: male vs female average score.

```python
plt.hist(df["G3"], bins = 10, color='skyblue', edgecolor='black')
plt.title("Histogram of grades")
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df["studytime"], df["G3"])
plt.title("study time vs grades")
plt.xlabel("study Time")
plt.ylabel("Grade")
plt.show()
```

```python
gen_avg.plot(kind='bar')
plt.title('male vs female average score')
plt.xlabel('Gender')
plt.ylabel('Average Grade')
plt.show()
```

**Conclusion**
